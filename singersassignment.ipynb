```python
import pandas as pd
```

```python
data = {
    'Singer': ['Arijit Singh', 'Shreya Ghoshal', 'Sonu Nigam', 'Lata Mangeshkar', 'Kishore Kumar'],
    'Years_Active': [15, 20, 30, 70, 40],
    'Number_of_Songs': [500, 700, 600, 1000, 800],
    'Awards_Won': [50, 60, 40, 100, 80]
}
```

```python
df = pd.DataFrame(data)
```

```python
df.info()
```

```python
#bar chart
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))
plt.bar(df['Singer'], df['Number_of_Songs'], color='lightblue')
plt.title('number of songs by indian singers')
plt.xlabel('singers')
plt.ylabel('number of songs')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
