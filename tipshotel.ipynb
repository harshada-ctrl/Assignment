```python
import pandas as pd
import numpy as np

# Sample dataset
data = {
    "Customer ID": np.arange(1, 31),
    "Total Bill": np.random.uniform(200, 1000, 30).round(2),
    "Tip Amount": np.random.uniform(10, 100, 30).round(2),
    "Smoking": np.random.choice(['Yes', 'No'], 30),
    "Date": pd.date_range(start='2023-01-01', periods=30, freq='D')
}

df = pd.DataFrame(data)
print(df.head(30))
df.to_csv('tips.csv', index=False)
```

```python
### Step 2: Represent the Relationship Between Total Bill Amount and Tip Amount Using Scatter Plot

import matplotlib.pyplot as plt

# Scatter plot
plt.scatter(df['Total Bill'], df['Tip Amount'])
plt.title('Total Bill vs Tip Amount')
plt.xlabel('Total Bill (Rs.)')
plt.ylabel('Tip Amount (Rs.)')
plt.grid(True)
plt.show()
```

```python
### Step 3: Draw a Graph Depicting Percentage of Smoking vs Non-Smoking Customers Using Pie Chart
# Pie chart
#sizes: The data for the slices, representing proportions of the whole.
#labels: The labels for each slice.
#autopct: Displays the percentage on each slice.
#startangle: Rotates the start of the pie chart to make it more visually appealing.
#This function is versatile and can be customized to create a pie chart that fits various visualization needs.
smoking_counts = df['Smoking'].value_counts()
plt.pie(smoking_counts, labels=smoking_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Percentage of Smoking vs Non-Smoking Customers')
plt.show()
```

```python
### Step 4: Draw a Line Chart of Total Bill for the First 10 Entries

# Line chart
df_first_10 = df.head(20)
plt.plot(df_first_10['Customer ID'], df_first_10['Total Bill'], marker='o')
plt.title('Total Bill for the First 10 Entries')
plt.xlabel('Customer ID')
plt.ylabel('Total Bill (Rs.)')
plt.grid(True)
plt.show()
```

```python
### Step 5: Plot the Total Bill Amount for the First 20 Entries Using Bar Graph

# Bar graph
df_first_20 = df.head(20)
plt.bar(df_first_20['Customer ID'], df_first_20['Total Bill'])
plt.title('Total Bill Amount for First 20 Entries')
plt.xlabel('Customer ID')
plt.ylabel('Total Bill (Rs.)')
plt.xticks(rotation=90)
plt.show()
```
